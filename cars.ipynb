```python
page_contents = response.text
```

# Use Beautiful Soup to parse and extract Information
- Parse and explore the structure of the website
- Create functions to extract from the page into list or dictionary

```python
!pip install beautifulsoup4 --upgrade --quiet
```

```python
from bs4 import BeautifulSoup
```

```python
doc = BeautifulSoup(page_contents,'html.parser')
```

```python
type(doc)
```

```python
car_name = doc.find_all('h2',{'class':'title'})
```

```python
car_name[3].text
```

```python
car_price = doc.find_all('span',{'class':"primary-price"})
```

```python
car_price[3].text
```

```python
car_names = []

for tag in car_name:
    car_names.append(tag.text)

print(car_names)
```

```python
car_prices = []

for tag in car_price:
    car_prices.append(tag.text.strip("[]"))
    
print(car_prices)
```

```python
!pip install pandas --quiet
```

```python
import pandas as pd
```

```python
dict = {
    'Car Name': car_names,
    'Prices': car_prices
}

df = pd.DataFrame(dict)
```

```python
df.to_csv('aston_matrina.csv')
```
